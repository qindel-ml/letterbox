```python
from PIL import Image
from AugmentedLetterbox import AugmentedLetterbox
```

```python
img = Image.open('Valencia_Sightseeing_0001_(203798303).jpeg')

num_rows = 10
num_cols = 2
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

fig, ax = plt.subplots(num_rows, num_cols, figsize=(15, 15 * num_rows // num_cols))

for row in range(num_rows):
    for col in range(num_cols):

        lbimg = AugmentedLetterbox(img)

        lbimg.letterbox(608, 608, augments={'hue':0.1, 'saturation':0.25, 'value':0.25, 'hflip_prob':0.3, 
                                    'vflip_prob':0.1, 'rotate_prob':0.3, 'rotate_angle':20, 'rotate_expand':1, 
                                    'jitter_prob':0.2, 'jitter':0.4, 'scale_prob':0.2, 'scale_min':0.1, 
                                    'scale_max':0.2, 'crop_prob':0.5, 'crop_frac':0.4}, randomize_pos=True)


        ax[row][col].imshow(lbimg)
```
